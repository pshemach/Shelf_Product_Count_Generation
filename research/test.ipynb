```python
import cv2
import numpy as np
import base64
import json
import os
```

```python
img = cv2.imread('data/test_images/cropped_image_3.jpg')
```

```python
def encode_image(image: np.ndarray) -> str:
    _, buffer = cv2.imencode('.jpg', image)
    return base64.b64encode(buffer).decode('utf-8')

encode_image(img)
```
